```python
import spacy
import pandas as pd
```

```python
df1 = pd.DataFrame(columns = ['Text', 'pos'])
df2 = pd.DataFrame(columns = ['Text', 'Label'])
```

```python
def annotate1(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    rows = []
    for token in doc:
        print(token.text, token.pos_)
        rows.append([token.text, token.pos_])
    df = pd.DataFrame(rows, columns=['Text', 'pos'])
    #df.to_csv('C:\Users\LENOVO\Desktop\cours\S5\Ingénierie cognitive\Projet\data.csv')
    return df
```

```python
annotate1('Apple is looking at buying U.K. startup for $1 billion')
```

```python
def annotate2(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    rows = []
    for ent in doc.ents:
        print(ent.text, ent.label_)
        rows.append([ent.text, ent.label_])
    df = pd.DataFrame(rows, columns=['Text', 'label'])
    #df.to_csv(r'C:\Users\LENOVO\Desktop\cours\S5\Ingénierie cognitive\Projet\data2.csv')
    return df
   
```
